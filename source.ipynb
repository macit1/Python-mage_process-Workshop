```python
# Load an image
image = cv2.imread(image_path_cats)

# Add random noise to the image
noise = np.random.normal(loc=0, scale=20, size=image.shape)
noisy_image = image + noise.astype('uint8')

# Apply a simple threshold to detect noise
threshold_value = 30
noise_mask = cv2.threshold(np.abs(image - noisy_image), threshold_value, 255, cv2.THRESH_BINARY)[1]

# Display the original image, noisy image, and the noise mask
cv2.imshow('Original Image', image)
cv2.imshow('Noisy Image', noisy_image)
cv2.imshow('Noise Mask', noise_mask)

# Wait for a key press and then close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(image)
axs[1].imshow(noisy_image)
axs[2].imshow(noise_mask)
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')
plt.show()
```

```python

# Define a video capture object
vidcap = cv2.VideoCapture('data/video_1.mp4')

# Set the window name
cv2.namedWindow("Video", cv2.WINDOW_NORMAL)

# Creating a loop for running the video and saving all the frames
while True:
    # Capture video frame by frame
    success, img_capt = vidcap.read()

    # Check if the video has ended
    if not success:
        print("Video Finished")
        break

    # Maintain aspect ratio while resizing
    target_width = 300
    aspect_ratio = img_capt.shape[1] / img_capt.shape[0]
    target_height = int(target_width / aspect_ratio)

    # Resize the image frames
    resize = cv2.resize(img_capt, (target_width, target_height))

    # Display the frames
    cv2.imshow("Video", resize)

    # Closing the video by Escape button
    if cv2.waitKey(10) == 27:
        print('Escaped')
        break

# Release video capture resources
vidcap.release()
cv2.destroyAllWindows()
```
